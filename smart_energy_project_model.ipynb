```python
import pandas as pd

def aggregate_data_to_hourly(df, time_col, power_cols, weather_cols):
    """
    Aggregates per-minute data into hourly data.

    Parameters:
    - df: DataFrame containing the data.
    - time_col: The column name that contains the timestamp.
    - power_cols: List of column names related to power consumption to be summed.
    - weather_cols: List of column names related to weather data to be averaged.

    Returns:
    - hourly_data: DataFrame with the data aggregated to hourly.
    """

    # Convert the specified 'time_col' to datetime format for resampling
    df[time_col] = pd.to_datetime(df[time_col])

    # Set the datetime column as the index
    df.set_index(time_col, inplace=True)

    # Create a new DataFrame for the aggregated data
    hourly_data = pd.DataFrame()

    # Sum power-related data and average weather-related data on an hourly basis
    hourly_data[power_cols] = df[power_cols].resample('H').sum()
    hourly_data[weather_cols] = df[weather_cols].resample('H').mean()

    return hourly_data

power_columns = ['House overall', 'Dishwasher', 'Home office', 'Fridge', 'Wine cellar', 'Garage door', 'Barn', 'Well', 'Microwave', 'Living room', 'Furnace', 'Kitchen']
weather_columns = ['temperature', 'humidity', 'visibility', 'pressure', 'windSpeed', 'cloudCover', 'windBearing', 'precipIntensity', 'dewPoint', 'precipProbability']
hourly_df = aggregate_data_to_hourly(df, 'start_time', power_columns, weather_columns)
```

#VAR MODEL

```python
import pandas as pd
from statsmodels.tsa.vector_ar.var_model import VAR

# Target and feature columns
target_columns = ['Dishwasher', 'Home office', 'Fridge', 'Wine cellar', 'Garage door', 'Barn', 'Well', 'Microwave', 'Living room', 'Furnace', 'Kitchen']
feature_columns = ['temperature', 'humidity', 'visibility', 'pressure', 'windSpeed', 'cloudCover', 'precipIntensity', 'dewPoint', 'precipProbability', 'month', 'day', 'weekday', 'weekofyear', 'hour']

# Spliting the data into 80% for training and 20% for testing
split_index = int(len(hourly_df) * 0.8)
train = hourly_df.iloc[:split_index]
test = hourly_df.iloc[split_index:]

# Fit the VAR model on the training data
model = VAR(endog=train[target_columns], exog=train[feature_columns])
model_fit = model.fit()

# Make prediction on the test data
prediction = model_fit.forecast(y=test[target_columns].values, steps=len(test), exog_future=test[feature_columns].values)

# Calculate absolute differences between forecasted and actual values
absolute_errors = np.abs(prediction - test[target_columns].values)

# Calculate mean absolute error
mae = np.mean(absolute_errors)
print("Mean Absolute Error (MAE):", mae)
```

#LSTM

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import mean_squared_error, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adagrad
from tensorflow.keras.regularizers import l1

# Standardizing the features (regressors) for PCA
scaler_features = StandardScaler()
features_scaled = scaler_features.fit_transform(hourly_df[regressors].fillna(0))

# Applying PCA to the features
pca = PCA(n_components=0.95)
features_pca = pca.fit_transform(features_scaled)

# Scale the targets
scaler_targets = MinMaxScaler(feature_range=(0, 1))
targets_scaled = scaler_targets.fit_transform(hourly_df[targets].fillna(0))

# Combine PCA features with scaled targets
features_pca_df = pd.DataFrame(features_pca, index=hourly_df.index)
targets_df = pd.DataFrame(targets_scaled, index=hourly_df.index, columns=targets)
combined_df = pd.concat([features_pca_df, targets_df], axis=1)

# Split data into training and testing based on 'month'
combined_df['month'] = hourly_df['month']
train = combined_df[combined_df['month'] < 11]
test = combined_df[combined_df['month'] >= 11]
train.drop(columns=['month'], inplace=True)
test.drop(columns=['month'], inplace=True)

# Define LSTM sequences
time_steps = 3
def create_sequences(data, time_steps=1):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps), :-len(targets)])
        y.append(data[i + time_steps, -len(targets):])
    return np.array(X), np.array(y)

X_train, y_train = create_sequences(train.values, time_steps)
X_test, y_test = create_sequences(test.values, time_steps)


# Model configuration
time_steps = 3  # The number of timesteps per input sample
n_features = X_train.shape[2]
n_outputs = y_train.shape[1]

model = Sequential([
    LSTM(100, activation='relu', input_shape=(time_steps, n_features), return_sequences=True),
    Dropout(0.3),
    LSTM(100, activation='relu'),
    Dropout(0.3),
    Dense(n_outputs, kernel_regularizer=l1(0.01)),  # L1 regularization
])

optimizer = Adagrad(learning_rate=0.001)

model.compile(optimizer=optimizer, loss='mean_absolute_error')

# Training the model
model.fit(X_train, y_train, epochs=50, batch_size=64, validation_split=0.1, verbose=1)

# Forecasting and Evaluation
def forecast_next_hours(model, X_seq, future_steps, n_features, n_targets):
    pred_seq = X_seq[-1:]
    predictions = []
    for _ in range(future_steps):
        pred = model.predict(pred_seq)
        predictions.append(pred[0])
        pred_seq = np.roll(pred_seq, -1, axis=1)
        pred_seq[0, -1, :n_features] = pred[0, :n_features]
    return np.array(predictions)

forecasted_test = forecast_next_hours(model, X_test, len(X_test), X_train.shape[2], y_train.shape[1])
forecasted_test = scaler_targets.inverse_transform(forecasted_test)

# Evaluation
true_values = scaler_targets.inverse_transform(y_test)
mse = mean_squared_error(true_values, forecasted_test)
mae = mean_absolute_error(true_values, forecasted_test)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```

#SARIMAX

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Target and regressor columns
targets = ['Dishwasher', 'Home office', 'Fridge', 'Wine cellar', 'Garage door', 'Barn', 'Well', 'Microwave', 'Living room', 'Furnace', 'Kitchen']
regressors = ['temperature', 'humidity', 'visibility', 'pressure', 'windSpeed', 'cloudCover', 'precipIntensity', 'dewPoint', 'precipProbability', 'day', 'hour', 'weekday', 'weekofyear']

# Creating a 'month' column from the one-hot encoded 'month_1', 'month_2', etc.
month_columns = [f'month_{i}' for i in range(1, 13)]
hourly_df['month'] = hourly_df[month_columns].idxmax(axis=1).str.extract('(\d+)').astype(int)

# Standardize the features
scaler_features = StandardScaler()
features_scaled = scaler_features.fit_transform(hourly_df[regressors].fillna(0))
features_scaled_df = pd.DataFrame(features_scaled, index=hourly_df.index, columns=regressors)

# Standardize the targets separately
scaler_targets = StandardScaler()
targets_scaled = scaler_targets.fit_transform(hourly_df[targets].fillna(0))
targets_scaled_df = pd.DataFrame(targets_scaled, index=hourly_df.index, columns=targets)

# Split the data into train and test based on the 'month' column
combined_df = pd.concat([features_scaled_df, targets_scaled_df], axis=1)
combined_df['month'] = hourly_df['month']

train = combined_df[combined_df['month'] < 11]
test = combined_df[combined_df['month'] >= 11]

train.drop(columns=['month'], inplace=True)
test.drop(columns=['month'], inplace=True)

# Train and forecast using SARIMAX for each target variable
forecasted_test = pd.DataFrame(index=test.index, columns=[f"{t}_predicted" for t in targets])

for target in targets:
    # Create and train the SARIMAX model for the current target
    model = SARIMAX(train[target], exog=train[regressors], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
    sarimax_fit = model.fit(disp=False)

    # Forecast on the test data
    forecast = sarimax_fit.predict(exog=test[regressors], start=test.index[0], end=test.index[-1])
    forecasted_test[f"{target}_predicted"] = forecast

# Inverse transform the scaled predictions to their original scale
forecasted_test_values = scaler_targets.inverse_transform(forecasted_test.values)

# Prepare the final DataFrame with the predicted values
final_forecast_df = pd.DataFrame(forecasted_test_values, index=test.index, columns=[f"{t}_predicted" for t in targets])

# Evaluation
true_values = scaler_targets.inverse_transform(test[targets])
mse = mean_squared_error(true_values, forecasted_test_values)
mae = mean_absolute_error(true_values, forecasted_test_values)

print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```
